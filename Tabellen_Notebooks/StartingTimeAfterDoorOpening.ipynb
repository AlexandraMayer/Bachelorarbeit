# Startzeiten der Aussteiger

Berechnung der Startzeiten für erste Aussteiger bezogen auf ihren Typ (aggressiv, defensiv, normal).
1. Laden der Daten
2. Datenvorverarbeitung: Zeiten für die Typen
3. Boxplot der Zeiten für Typen
4. Histogramme der Zeiten für Typen

## Bibliotheken

```python
import pandas as pd

import matplotlib.pyplot as plt
```

## Stildefintion für Plots

```python
#Definierter Plotstyle für die Bachelorarbeit
#Muss zu den anderen Bibliotheken hinzugefügt werden
#Datei in Dateien eingefügt
plt.style.use("ba")
```

## Helferfunktionen

```python
def_dic = {'defensive': -1, 'normal': 0, 'aggressive': 1}

def load_data(path):
    return pd.DataFrame(pd.read_csv(path+file_suffix, sep=";"))

def get_time_for_type(data, type_):
    data_type = data.where(data['Type'] == def_dic[type_]).dropna()
    return data_type['Time']
```

## Laden der Daten

```python
path = "Tabels/TimeAfterDoorOpening"
file_suffix = ".csv"
```

```python
data_start_time = load_data(path)
#data_start_time
```

# Datenvorverarbeitung

```python
time_def = get_time_for_type(data_start_time, 'defensive').drop(3)
time_norm = get_time_for_type(data_start_time, 'normal').drop(8).drop(17)
time_ag = get_time_for_type(data_start_time, 'aggressive').drop(5)
```

## Boxplot der Startzeiten

```python
plt.figure(figsize=(16, 12))
plt.boxplot([time_def, time_norm, time_ag])
plt.xticks([1, 2, 3], ['defensiv', 'normal', 'aggressiv'])
plt.plot(1, time_def.mean(), 'o', ms=4, label="Mittelwert Zeit defensiv: {:.1f} s".format(time_def.mean()))
plt.plot(2, time_norm.mean(), 'o', ms=4, label="Mittelwert Zeit normal:  {:.1f} s".format(time_norm.mean()))
plt.plot(3, time_ag.mean(), 'o', ms=4, label="Mittelwert Zeit aggressiv: {:.1f} s".format(time_ag.mean()))
plt.xlabel('Typen')
plt.ylabel('Startzeit')
plt.legend()
plt.savefig('..\\Text\\pictures\\data_evaluation\\types\\starting_time\\starting_times.png')
plt.show()
print("Normale Intervall [oberes Quartil, unteres Quartil] = [{:.1f}, {:.1f}]".format(time_norm.quantile(.25), time_norm.quantile(.75)))
print("Defensive Intervall [oberes Quartil, unteres Quartil] = [{:.1f}, {:.1f}]".format(time_def.quantile(.25), time_def.quantile(.75)))
print("Agressive Intervall [oberes Quartil, unteres Quartil] = [{:.1f}, {:.1f}]".format(time_ag.quantile(.25), time_ag.quantile(.75)))
```
